```python
# Recommender System 1 Uses Plot to generate Recommendations
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
def generate_vectors_tfidf(df):
    tfidf = TfidfVectorizer(stop_words='english')
    df['plot'] = df['plot'].fillna('')
    tfidf_matrix = tfidf.fit_transform(df['plot'])
    return tfidf_matrix
```

```python
def cosine_similarity(matrix,df):
    cosine_sim = linear_kernel(matrix,matrix)
    indices = pd.Series(df.index,index = df['title']).drop_duplicates()
    return cosine_sim,indices
```

```python
def get_recommendations(title,cosine_sim,indices,df):
    index = indices[title]
    sim_scores = [(i,val) for i,val in enumerate(cosine_sim[index])]
    sim_scores = sorted(sim_scores,key=lambda x: x[1],reverse = True)[1:9]
    movie_index = [tupel[0] for tupel in sim_scores]
    return df['title'].iloc[movie_index]
```

```python
#Testing the Functions
```

```python
def tester_func(title,df):
    vectors = generate_vectors_tfidf(df)
    cosine_simi,indices = cosine_similarity(vectors,df)
    recommendations = get_recommendations(title,cosine_simi,indices,df)
    return recommendations
```

```python
tester_func('The Dark Knight Rises',df)
```

```python
# Creating Recommendations Generator Function
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def count_vectorize(df):
    count = CountVectorizer(stop_words = 'english')
    count_matrix = count.fit_transform(df['word_soup'])
    return count_matrix
```

```python
def cosine_similarity_2(matrix,df):
    cosine_sim = cosine_similarity(matrix,matrix)
    dataframe = df.reset_index()
    index = pd.Series(dataframe.index,index = dataframe['title'])
    return cosine_sim,index
```

```python
def get_recommendations_2(title,cosine_sim,indices,df):
    index = indices[title]
    sim_scores = cosine_sim[index]
    sim_scores = sorted([(idx,val) for idx,val in enumerate(sim_scores)],key=lambda x: x[1], reverse = True)[1:9]
    movie_index = [tupel[0] for tupel in sim_scores]
    return df['title'].iloc[movie_index]
```

```python
#Testing the Functions
```

```python
def tester_func_2(title,df):
    vectors = count_vectorize(df)
    cosine_simi,indices = cosine_similarity_2(vectors,df)
    recommendations = get_recommendations_2(title,cosine_simi,indices,df)
    return recommendations
```

```python
tester_func_2('The Dark Knight Rises',df)
```
